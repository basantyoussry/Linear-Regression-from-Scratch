```python

# importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# importing or loading the dataset
df = pd.read_csv('/content/Salary.csv')
print(df)
df.info()
# df.shape
# distributing the dataset into two components X and Y
X = df.iloc[:, 0].values
y = df.iloc[:, 1:2].values
```

```python
def model(a,b):
    y=a*X+b
    return y
    # print(y)
model(2,3)
import matplotlib.pyplot as plt
plt.plot(X, y,'o')
a, b = np.polyfit(X, y, 1)
plt.plot(X, model(13600,5))
plt.plot(X,a*X+b)
```

```python
import random
def mse():
  y_pred = []
  for i in range(0,30):
    n = random.randint(2000,120000)
    y_pred.append(n)
  array1 = np.array(y)
  array2 = np.array(y_pred)
  n=len(df)
  for i in range (0,n):
    MSE=np.square(np.subtract(y,y_pred))/len(df)
    plt.plot(X,y)
    plt.plot(X,y_pred)
    plt.title('SActual Values Of Y"Salary" + Predictions')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(['y_actual','y_pred'])
    plt.show()
    return MSE
print( "The Mean Square Error is: " ,mse())
```
