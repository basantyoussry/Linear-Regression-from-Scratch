```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

all_losses = []
# Load the salary data
# data = pd.read_csv('salary.csv')

# Extract the feature and target variables
# X = data['years_of_experience'].values
# Y = data['salary'].values

# Define the linear equation
def linear_eqn(theta0, theta1, x):
    return theta0 + theta1 * x

# Define the loss function
def mse_loss(Y, Y_hat):
    return np.mean((Y - Y_hat) ** 2)

# Define the gradient descent function
def gradient_descent(X, Y, theta0, theta1, alpha, epochs):
    m = len(X)
    for i in range(epochs):
        Y_hat = linear_eqn(theta0, theta1, X)
        loss = mse_loss(Y, Y_hat)
        all_losses.append(loss)
        #  d_theta = (-2/m) * np.dot(X.T, (Y - Y_hat))
        d_theta0 = (-1/m) * np.sum(Y - Y_hat)
        d_theta1 = (-1/m) * np.dot(X.T, (Y - Y_hat))

        theta0 = theta0 - alpha * d_theta0
        theta1 = theta1 - alpha * d_theta1
    return theta0, theta1

# Set the learning rate and number of epochs
alpha = 0.3
epochs = 1000

# Initialize the parameters (theta0 and theta1)
theta0 = 0
theta1 = 0

# Perform gradient descent to find optimal parameters
theta0, theta1 = gradient_descent(X, Y, theta0, theta1, alpha, epochs)

# Print the optimal parameters
print("Optimal parameters: ", theta0, theta1)

# Plot the data and the linear regression line
plt.scatter(X, Y)
plt.plot(X, linear_eqn(theta0, theta1, X), color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
